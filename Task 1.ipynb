```python
## Importing Required Libarires 

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

```

```python
## Load data from CSV file

df = pd.read_csv ('Data.csv')
df
```

```python
## Display Initial Data Info 

print(df.info())
```

```python
## Clean the data by dropping rows with missing values in 'IncomeGroup'

df_cleaned = df.dropna(subset=['IncomeGroup'])
df
```

```python
## Bar chart for the categorical variable 'IncomeGroup'

plt.figure(figsize=(10, 5))
sns.countplot(x='IncomeGroup', data=df_cleaned, palette='viridis')
plt.title('Bar Chart of Income Group')
plt.xlabel('Income Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
## Histogram for the numerical variable 'RandomValues'

plt.figure(figsize=(10, 5))
sns.histplot(df_cleaned['RandomValues'], bins=10, kde=True, color='blue')
plt.title('Histogram of Random Values')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```
